```python
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
import numpy as np
import sklearn
```

```python
X,y_true=make_blobs(n_samples=300,centers=4,cluster_std=0.6,random_state=0)
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmean=KMeans(n_clusters=4)
kmean.fit(X)
y_kmean=kmean.predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=y_kmean,s=10)
centers=kmean.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c="black",s=200)
```

```python
from sklearn.metrics import pairwise_distances_argmin
```

```python
from sklearn.metrics import pairwise_distances_argmin
def find_clusters(X, n_clusters, rseed=2):
    # 1. Randomly choose clusters
    rng = np.random.RandomState(rseed)
    i = rng.permutation(X.shape[0])[:n_clusters]
    centers = X[i]
    print("centers",centers)
    
    while True:
        # 2a. Assign labels based on closest center
        labels = pairwise_distances_argmin(X, centers)
        print("lable",labels)
        
        # 2b. Find new centers from means of points
        new_centers = np.array([X[labels == i].mean(0) for i in range(n_clusters)])
        print("new_centers",new_centers)
        
        # 2c. Check for convergence
        if np.all(centers == new_centers):
            break
        centers = new_centers
    return centers, labels
```

```python
centers, labels = find_clusters(X, 4)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels,
s=50, cmap='viridis');
```

```python
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
#digits co nhungn feature: data, target_names,target
```

```python
kmean=KMeans(n_clusters=10,random_state=0)
clusters=kmean.fit_predict(digits.data)
```

```python
print(clusters)
```

```python
kmean.cluster_centers_.shape
```

```python
fig,ax=plt.subplots(2,5,figsize=(8,3))
centers=kmean.cluster_centers_.reshape(-1,8,8)
for axi,i in zip(ax.flat,centers):
    axi.set(xticks=[],yticks=[])
    axi.imshow(i,cmap=plt.cm.binary,interpolation='nearest',)
```

Ta can sap xep lai thu tu tu 0-9 cho cac label

```python
from scipy.stats import mode
labels = np.zeros_like(clusters)
for i in range(10):
    mask = (clusters == i)
    labels[mask] = mode(digits.target[mask])[0]
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
mat = confusion_matrix(digits.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
xticklabels=digits.target_names,
yticklabels=digits.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

```python
accuracy_score(digits.target,labels)
```

```python
china=load_sample_image("china.jpg")
```

```python
ax=plt.axes(xticks=[],yticks=[])
ax.imshow(china)
```

```python
china.shape
```

```python
# Normilize pixel ve 0-1
data=china/255.0
data=data.reshape(427*640,3)
```

```python
#ve cac diem mau len khong gian 2 chieu
def plot_pixels(data,title,colors=None,N=1000):
    if colors is None:
        colors=data
    rs=np.random.RandomState(0)
    #hoan vi N pixel dau tien cua mang
    i=rs.permutation(data.shape[0])[:N]
    colors=colors[i]
    R,G,B=data[i].T
    
    fig,ax=plt.subplots(1,2,figsize=(16,6))
    ax[0].scatter(R,G,c=colors,marker=".")
    ax[0].set(xlabel="Red",ylabel="Green",xlim=(0,1),ylim=(0,1))
    ax[1].scatter(R,B,c=colors,marker=".")
    ax[1].set(xlabel="Red",ylabel="Blue",xlim=(0,1),ylim=(0,1))
    fig.suptitle(title,size=20);
              
```

```python
plot_pixels(data,"1000 diem anh")
```

```python
#ta se giam tu 16 trieu mau ve 16 mau, ta chi giam mau chu khong giam chieu
from sklearn.cluster import MiniBatchKMeans
kmeans=MiniBatchKMeans(16)
kmeans.fit(data)
new_colors=kmeans.cluster_centers_[kmeans.predict(data)]
plot_pixels(data,colors=new_colors,title="16 mau")
```
